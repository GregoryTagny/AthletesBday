## II - Getting the link of all major leagues per country

Creating a function to extract all the leagues per country

```python
def ExtractMajorLeague(country_id):
    '''
        Extracts top 3 championship in the country
        Input: {country_id}
        Output: list of max 3 tuples(league_name|str, link|str)
    '''
    major_leagues = []
    # Making a request to page link
    page = requests.get(url_countries+str(country_id), headers={'User-Agent': 'Custom'})
    soup = BeautifulSoup(page.text, 'html.parser')
    
    # Extracting the top 3 leagues in the country
    for i in range(3):
        league_division = i+1
        try:
            league_block = soup.find(class_=["items"]).find_all(class_="inline-table")[i].find_all('a')[-1]
            league_name, league_link = (league_block.text, homepage+league_block['href'])
            major_leagues.append((league_name, league_division, league_link))
        except:
            major_leagues.append((np.NaN,np.NaN,np.NaN))
            
    return major_leagues    
```

Creating a for loop over all the countries

```python
major_leagues_all_countries = []
for id in all_countries_df['Page Code']:
    major_leagues_all_countries.append(ExtractMajorLeague(id))
major_leagues_all_countries
```

```python
#Reset df in case of duplicates issues
all_countries_df = all_countries_df[['Country','Page Code']].drop_duplicates().reset_index(drop=True).sort_values(by=['Country'])
```

```python
# Replicating the df 3 times for each major league
all_leagues_df = pd.concat([all_countries_df]*3, ignore_index=True).sort_values(by=['Country']).reset_index(drop=True)
all_leagues_df.head()
```

```python
# Turning all the tupples into list 
league_name_list = []
league_link_list = []
league_division_list = []
for league in major_leagues_all_countries:
        for elt in league:                
            league_name_list.append(elt[0])
            league_division_list.append(elt[1])
            league_link_list.append(elt[2])
```

```python
# Creating new columns
all_leagues_df['League Name'] = league_name_list
all_leagues_df['League Division'] = league_division_list
all_leagues_df['Link'] = league_link_list

# Filtering only the valid leagues
all_leagues_df = all_leagues_df.dropna().reset_index(drop=True)
all_leagues_df.to_csv('All Leagues.csv')
all_leagues_df.head(10)
```

## III - Extracting the link for all clubs

```python
def ExtractClubs(league_link):
    '''
        Extracts all teams in the championship
        Input: {league_link}|str
        Output: list of tuples(club_name|str, link|str)
    '''
    all_clubs = []
    # Making a request to page link
    page = requests.get(league_link, headers={'User-Agent': 'Custom'})
    soup = BeautifulSoup(page.text, 'html.parser')
    
    # Extracting all clubs
    club_session = soup.find(class_=["items"]).find_all(class_="hauptlink")
    club_session = [elt.find('a') for elt in club_session]
    for i in range(0, len(club_session),2):
        club_name, club_link = (club_session[i].text, homepage+club_session[i]['href'])
        all_clubs.append((club_name, club_link))
        
    return all_clubs    
```

Creating a list of all the clubs with their links

```python
all_club_list = []
# Retrieving each row of the df
for index, row in all_leagues_df.iterrows():
    try:
        # For each row, extracting the link 
        for club in ExtractClubs(row[-1]):
            # Creating a tuple for each team
            all_club_list.append((row[0]
                                 ,row[1]
                                 ,row[2]
                                 ,row[3]
                                 ,row[4]
                                 ,club[0]
                                 ,club[1]
            ))
    except:
        all_club_list.append((row[0]
                     ,row[1]
                     ,row[2]
                     ,row[3]
                     ,row[4]
                     ,np.NaN
                     ,np.NaN
        ))

all_club_list
```

```python
all_clubs_df = pd.DataFrame(columns=['Country','Page Code','League Name','League Division', 'League link', 'Club Name', 'Club Link']
                                     , data=all_club_list)
all_clubs_df.to_csv('All clubs.csv')
all_clubs_df.tail(10)
```

## IV - Getting all the players' bdays

```python
def ExtractAllBdays(club_page_link):
    '''
        Extracts all bdays on the current team's page
        Input: {team_link}|str
        Output: list of tuples(player_name|str, bday|str)
    '''
    
    players_list = []
    # Making a request to page link
    page = requests.get(club_page_link, headers={'User-Agent': 'Custom'})
    soup = BeautifulSoup(page.text, 'html.parser')
    
    # Extracting all players
    players_session = soup.find(class_=["items"]).find_all(class_=["odd"]) + soup.find(class_=["items"]).find_all(class_=["even"])
    for player in players_session:
        # Extracting all attributes for a given player
        attributes = player.find_all("td")
        position = attributes[4].text
        player_name = attributes[5].text
        bday = attributes[6].text.split(' (')[0]
        nationality = attributes[7].find('img')['alt']
        try:
            player_value = attributes[8].text.split()[0]
        except:
            player_value = np.NaN

        players_list.append((player_name, position, bday, nationality,player_value))

    return players_list
```

```python
all_players_list = []

# Retrieving each row of the df
for index, row in all_clubs_df.iterrows():
    try:
        # For each row, extracting the link 
        for player in ExtractAllBdays(row[-1]):
            # Creating a tuple for each team
            all_players_list.append((row[0]
                                 ,row[1]
                                 ,row[2]
                                 ,row[3]
                                 ,row[4]
                                 ,row[5]
                                 ,row[6]
                                 ,player[0]
                                 ,player[1]
                                 ,player[2]
                                 ,player[3]
                                 ,player[4]
                                  
            ))
    except:
        # Creating a tuple for each team
        all_players_list.append((row[0]
                             ,row[1]
                             ,row[2]
                             ,row[3]
                             ,row[4]
                             ,row[5]
                             ,row[6]
                             ,player[0]
                             ,player[1]
                             ,player[2]
                             ,player[3]
                             ,player[4]

        ))

all_players_list
```

```python
all_players_df = pd.DataFrame(columns=['Country','Page Code','League Name','League Division'
                                       ,'League link', 'Club Name', 'Club Link', 'Player Name'
                                       , 'Player Position', 'Birthday', 'Player Origin', 'Player Value']
                                     , data=all_players_list)
all_players_df.to_csv('All Players.csv')
all_players_df.tail(10)
```
